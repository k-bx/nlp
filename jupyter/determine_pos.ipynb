```python
import spacy
from spacy import displacy
nlp = spacy.load('en_core_web_lg')
```

```python
doc = nlp("We can but hope that everything will be fine.")
for t in doc:
    print(t.text, t.lemma_, t.pos_, t.tag_, t.dep_, t.shape_, t.is_alpha, t.is_stop)
displacy.render(doc, style='dep', jupyter=True, options=dict(compact=True))
```
